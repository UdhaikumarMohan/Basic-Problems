```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("water.csv")
df.head()
```

```python
df.info()
```

```python
ax = df.plot()
```

```python
top_ten  = df.head(10)
ax = top_ten.plot.bar(title='water',x='datetime',y='temp')
```

```python
plot_data =df.set_index(pd.to_datetime(df['datetime']))
```

```python
plot_data.drop('datetime', axis=1,inplace=True)
plot_data.head()
```

```python
ax = plot_data['height'].plot()
```

```python
plot_data.index.name = 'Date/time'
plot_data.rename(columns={'height':'Height(cm)'},inplace=True)
plot_data.head()

ax = plot_data['Height(cm)'].plot(title='TIDE',legend=True)
```
