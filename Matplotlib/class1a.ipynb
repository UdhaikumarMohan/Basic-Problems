```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
grades = pd.DataFrame([[5,4,8],[9,5,3],[8,4,7],[9,8,9],[2,9,3]],index=('Ram','Raj','Dev','Krish','Sri'),columns=('test_1','test_2','test_3'))
grades
```

Average Grade obtained in each subjects

```python
grades.mean()
```

```python
ax = grades.mean().plot()
```

```python
ax = grades.mean().plot.bar(title='Marks Scored')
```

```python
ax = grades.mean().plot.barh(title='Marks Scored')
```

ax = grades.mean(axis=1).plot.bar(title='Mark Scored')

```python
grades['Average']=grades.mean(axis=1)
grades
```

```python
ax = grades.plot.barh(title="All the grades")
```

Average of Class

```python
grades.loc['Whole class']=grades.mean()
grades
```

```python
ax = grades.plot.barh(title='Class Test')
```

Progress of each students

```python
grades_t = grades.T
grades_t
```

```python
grades_t.drop(["Whole class"],axis=1,inplace=True)
grades_t.drop('Average',inplace=True)
grades_t
```

```python
ax = grades_t.plot(title = "Student Programming")
```
