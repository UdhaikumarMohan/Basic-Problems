Scatter Plots

```python
%matplotlib notebook
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('athletes.csv')
df.head()
```

```python
df.info()
```

```python
f = df[df['sex']=='female']
m = df[df['sex']=='male']

tennis_f = f[f['sport']=='tennis']
tennis_m = m[m['sport']=='tennis']

gymnast_f = f[f['sport']=='gymnastics']
gymnast_m = m[m['sport']=='gymnastics']
```

```python
gymnast_f.plot.scatter(x='weight',y='height')

fig, ax = plt.subplots()

plots=[(tennis_f,'red','Tennis,women'),(tennis_m,'blue','Tennis,men'),(gymnast_f,'orange','Gymnastics,women'),(gymnast_m,'green','Gymnastics,men')]

for df,color,label in plots:
    
    df.plot.scatter(ax=ax,x='weight',y='height',alpha=.25,color=color,label=label)
    
ax.set_xlabel('Weight(kg)')
ax.set_ylabel('Height(cm)')
ax.set_title('Height and Weights of olympic Athletes');
    
fig, ax = plt.subplots()
gymnast_m.plot.scatter(ax=ax,x='height',y='weight',c='gold')
    
fig.tight_layout
```

```python
fig, ax = plt.subplots()

gymnast_m.plot.scatter(ax=ax,x='height',y='weight',c='gold',colormap='plasma')

fig.tight_layout()
```
